# 2. Limpieza

### Transformación de los datos de fecha

```python
df_75_26['fecha'] = pd.to_datetime(df_75_26['fecha'])
df_75_26['año'] = df_75_26['fecha'].dt.year
df_75_26['trimestre'] = df_75_26['fecha'].dt.quarter
df_75_26['periodo'] = df_75_26['año'].astype(str) + 'Q' + df_75_26['trimestre'].astype(str)

df_75_26 = df_75_26.set_index('fecha').sort_index()
df_75_26 = df_75_26.reset_index()

print(f'Información acerca de {df_75_26.attrs['nombre']}')
print(df_75_26.info())
```

### Eliminando la columna del indice de precios para suministro de agua, actividades de saneamiento, gestión de residuos y descontaminación,

Solo cuenta con datos a partir de 20073

```python
df_75_26 = df_75_26.drop(columns= ['IPRI_suministro_agua_actividades_saneamiento_gestion_residuos_descontaminacion_2021'])
```

```python
df_75_26.info()
```

### Cambio de base a 2025

```python
fechas = [2025]
columnas_temporales = ['fecha', 'año', 'trimestre', 'periodo']

columnas_ipri = [col for col in df_75_26.columns if col not in columnas_temporales]

# Calcular medias y conversores para cada columna IPRI
for columna in columnas_ipri:
    for año in fechas:
        datos_año = df_75_26[df_75_26['año'] == año][columna]

        if len(datos_año) == 0:
            print(f'  ✗ {columna} | Año {año}: sin datos')
            continue

        media = datos_año.mean()
        n_meses = len(datos_año)
        completitud = 'año completo' if n_meses == 12 else f'{n_meses}/12 meses'

        # Conversor: para llevar valores nominales a precios del año base
        col_conv = columna.replace('_2021', f'_conversor_{año}')
        df_75_26[col_conv] = media / df_75_26[columna]

        # Índice rebasado: base año = 100
        col_rebased = columna.replace('_2021', f'_{año}')
        df_75_26[col_rebased] = (df_75_26[columna] / media) * 100

        print(f'  ✓ {columna} | {año}: media = {media:.4f} ({completitud})')

print()
print('Columnas nuevas:')
print([col for col in df_75_26.columns if col not in columnas_temporales + columnas_ipri])
```

```python
df_75_26 = df_75_26.filter(regex='^(?!.*conversor)')
orden = ['fecha', 'año', 'trimestre', 'periodo'] + [columna for columna in df_75_26.columns if columna.startswith('IPRI_')]
df_75_26 = df_75_26[orden]
df_75_26
```

# 4. Guardando el dataset final en un archivo

```python
# ── Recortar al rango de tasa de paro y guardar en Datasets ────────────
from pathlib import Path
FECHA_INICIO = '1974-07-01'
FECHA_FIN    = '2025-10-01'
ruta_datasets = Path(r'C:\Users\marco\PycharmProjects\TFM_Marcos\Datasets')

df_final = df_75_26.copy()


df_75_26 = df_75_26[(df_75_26['fecha'] >= FECHA_INICIO) & (df_75_26['fecha'] <= FECHA_FIN)].copy()
df_75_26.to_csv(ruta_datasets / 'indice_precios_industriales_IPRI_bases_2021_2025.csv',
                index=False,
                encoding='utf-8-sig')
```
