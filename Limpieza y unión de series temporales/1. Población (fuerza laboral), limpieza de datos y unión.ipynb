# 2. Limpieza

### Transformación del tipo de dato y limpieza inicial

```python
# Conversión de las variables categóricas a "category" por eficiencia
variables_categoricas = ['Sexo', 'Grupos de edad (4)', 'Edad','Relación con la actividad económica']

for i, df in enumerate(dataframes):
    for variable in variables_categoricas:
        if variable in df.columns:
            df[variable] = df[variable].astype('category')
        else:
            continue
    dataframes[i] = df
```

### Conversión de los datos numéricos a formato internacional, eliminar puntos y reemplazar las comas "," por puntos "."

```python
for i in range(len(dataframes)):
    df = dataframes[i]

    # Convirtiendo a texto para poder manipularlo
    valores_limpios = df['Total'].astype(str).str.strip()

    # Eliminando los espacios internos y espacios irrompibles (\xa0) que usa el INE
    valores_limpios = valores_limpios.str.replace(' ', '', regex=False)
    valores_limpios = valores_limpios.str.replace('\xa0', '', regex=False)

    # Eliminando el punto de miles y cambiando la coma por punto
    valores_limpios = valores_limpios.str.replace('.', '', regex=False).str.replace(',', '.', regex=False)

    # Convirtiendo a numérico a numérico
    df['Total'] = pd.to_numeric(valores_limpios, errors='coerce')

    # Rellenando los nulos con 0.0 (valores que el INE guardaba con '..'
    df['Total'] = df['Total'].fillna(0.0)

    # Actualizamos la lista
    dataframes[i] = df

# Desempaquetamos fuera del bucle para no perder las referencias
df_76_95, df_96_01, df_02_25 = dataframes
```

```python
# Información acerca del tipo de dato y valores nulos de las columnas de cada dataset
for df in dataframes:
    print()
    print(f'Información acerca de {df.attrs['nombre']}')
    print(df.info())
```

### Transformación de los rangos de edad en las series temporales más recientes (2002 - 2025) para hgomogeneizar el dataset

```python
categorias_a_unificar = [
    'De 55 a 59 años',
    'De 60 a 64 años',
    'De 65 a 69 años',
    '70 y más años'
]

df_02_25['Edad'] = df_02_25['Edad'].astype(str).str.strip()

columnas_agrupacion = ['Sexo', 'Relación con la actividad económica', 'Periodo']
for col in columnas_agrupacion:
    df_02_25[col] = df_02_25[col].astype(str)

# Creando la máscara para filtrar
mask = df_02_25['Edad'].isin(categorias_a_unificar)

# Agrupando los datos filtrados y sumamos el 'Total'
df = df_02_25[mask].groupby(columnas_agrupacion, as_index=False, observed=True)['Total'].sum(min_count=0)

# Creando la columna 'edad' con el nombre unificado
df['Edad'] = 'De 55 y más años'

# Eliminando las categorías no agrupadas
df_02_25 = df_02_25[~mask]

# Unimos la nueva categoría al dataset original del que provenían los datos
df_02_25 = pd.concat([df_02_25, df], ignore_index=True)

df_02_25['Edad'] = df_02_25['Edad'].astype('category')

for col in columnas_agrupacion:
    df_02_25[col] = df_02_25[col].astype('category')

dataframes[2] = df_02_25
```

### Transformación para homogeneizar con la única serie retropolada del INE (parados)

```python
categorias_mas_25 = [
    'de 25 a 29 años',
    'de 30 a 34 años',
    'de 35 a 39 años',
    'de 40 a 44 años',
    'de 45 a 49 años',
    'de 50 a 54 años',
    'de 55 y más años',
    'de 25 a 54 años'
]


categorias_menos_25 = [
    'de 16 a 19 años',
    'de 20 a 24 años'
]
columnas_edad = ['Grupos de edad (4)', 'Edad']
for i, df in enumerate(dataframes):
    for columna in columnas_edad:
        if columna in df.columns:
            # Normalizar edad y columnas de agrupación a string/minúsculas ANTES de filtrar
            df[columna] = df[columna].astype(str).str.strip().str.lower()
            for col in columnas_agrupacion:
                df[col] = df[col].astype(str)

            # Creando la máscara para filtrar
            mask_mas = df[columna].isin(categorias_mas_25)
            mask_menos = df[columna].isin(categorias_menos_25)

            # Agrupando los datos filtrados y sumamos el 'Total'
            df_mas = df[mask_mas].groupby(columnas_agrupacion, as_index=False, observed=True)['Total'].sum(min_count=0)
            df_menos = df[mask_menos].groupby(columnas_agrupacion, as_index=False, observed=True)['Total'].sum(min_count=0)

            # Creando la columna 'edad' con el nombre unificado
            df_mas[columna] = '25_o_mas'
            df_menos[columna] = 'menos_de_25'

            # Eliminando las categorías no agrupadas
            df = df[~(mask_mas | mask_menos)]

            # Unimos la nueva categoría al dataset original del que provenían los datos
            partes = [d for d in [df, df_mas, df_menos] if len(d) > 0]
            df = pd.concat(partes, ignore_index=True)
            df[columna] = df[columna].astype(str).str.lower()
            df[columna] = df[columna].astype('category')

            for col in columnas_agrupacion:
                df[col] = df[col].astype('category')

            # Actualizamos la lista
            dataframes[i] = df

# Desempaquetamos fuera del bucle para no perder las referencias
df_76_95, df_96_01, df_02_25 = dataframes
```

### Transformación de los datos de "Periodo"

```python
trimestres_map = {
    'I': '1',
    'II': '2',
    'III': '3',
    'IV': '4',
    '1': '1', '2': '2', '3': '3', '4': '4'
}

# Conversión de "Periodo" a string y a formato internacional para manejarlo mejor
for i, df in enumerate(dataframes):
    df['Periodo'] = df['Periodo'].astype('string')

    # Separación de la información en diferentes columnas para su posterior uso
    df['año'] = df['Periodo'].str[:4]
    df['trimestre'] = df['Periodo'].str[5:].map(trimestres_map)

    texto_fecha = df['año'].astype('string') + "Q" + df['trimestre'].astype('string')
    df['Periodo'] = texto_fecha.astype('category')
    df['fecha'] = pd.PeriodIndex(texto_fecha, freq="Q").to_timestamp()

    df['año'] = df['año'].astype(int)
    df['trimestre'] = df['trimestre'].astype(int)

    df = df.set_index('fecha').sort_index()
    df = df.reset_index()
    dataframes[i] = df

    print(f'Información acerca de {df.attrs['nombre']}')
    print(df.info())

df_76_95, df_96_01, df_02_25 = dataframes
```

# 3. Unión en un único dataset

```python
df_final= pd.concat([df_76_95, df_96_01, df_02_25], ignore_index=True)
# Eliminar duplicados de ruptura metodológica 2002, conservar serie más moderna
cols_clave = [c for c in df_final.columns if c not in ['Total', 'total', 'valor', 'Value']]
n_antes = len(df_final)
df_final = df_final.drop_duplicates(subset=cols_clave, keep='last').reset_index(drop=True)
print(f'Filas: {n_antes} → {len(df_final)} ({n_antes - len(df_final)} duplicados eliminados)')

sexo_map = {
    'Ambos sexos': 'ambos_sexos',
    'Varones': 'hombres',
    'Hombres': 'hombres',
    'hombres': 'hombres',
    'Mujeres': 'mujeres',
    'mujeres': 'mujeres'
}

relacion_actividad_economica_map = {
    'Población de 16 y más años': 'poblacion_mayor_16',
    'Activos': 'activos',
    'Ocupados': 'ocupados',
    'Parados': 'parados',
    'Parados que buscan primer empleo': 'parados_1er_empleo',
    'Inactivos': 'inactivos',
    'Total': 'poblacion_mayor_16'
}


df_final['relacion_actividad_economica'] = df_final['relacion_actividad_economica'].map(relacion_actividad_economica_map)

df_final['sexo'] = df_final['sexo'].map(sexo_map)

columnas_texto = df_final.select_dtypes(include=['category']).columns

for columna in columnas_texto:
    df_final[columna] = df_final[columna].str.lower()

df_final = df_final.sort_values('fecha').reset_index(drop=True)

df_final
```

El cambio de metodología endureció las condiciones bajo las que se puede considerar a a una persona desempleada al considerar nuevos parámetros de búsqueda activa. En la nueva metodología todos esos parados que ahora no pasan el corte ahora son clasificados como inactivos.



```python
from pathlib import Path
ruta_parados = Path(r'C:\Users\marco\PycharmProjects\TFM_Marcos\Datasets') / 'parados_sexo_edad.csv'
df_paro = pd.read_csv(ruta_parados)
df_paro.head()
```

```python
# CHECKS DE CONSISTENCIA
# ocupados + parados = activos
# activos + inactivos = población_mayor_16

categorias = [
    'ambos_sexos_total', 'ambos_sexos_25_o_mas', 'ambos_sexos_menos_de_25',
    'hombres_total', 'hombres_25_o_mas', 'hombres_menos_de_25',
    'mujeres_total', 'mujeres_25_o_mas', 'mujeres_menos_de_25',
]

print('=== CHECK 1: ocupados + parados = activos ===')
for cat in categorias:
    suma = df_final[f'ocupados_{cat}'] + df_final[f'parados_{cat}']
    diff = (suma - df_final[f'activos_{cat}']).abs()
    print(f'  {cat}: error medio = {diff.mean():.2f} | máx = {diff.max():.2f}')

print()
print('=== CHECK 2: activos + inactivos = población_mayor_16 ===')
for cat in categorias:
    suma = df_final[f'activos_{cat}'] + df_final[f'inactivos_{cat}']
    diff = (suma - df_final[f'poblacion_mayor_16_{cat}']).abs()
    print(f'  {cat}: error medio = {diff.mean():.2f} | máx = {diff.max():.2f}')
```

```python
# CHECK FINAL COMBINADO
print('=== CHECK 3: ocupados + parados + inactivos = población_mayor_16 ===')
for cat in categorias:
    suma = (df_final[f'ocupados_{cat}']
          + df_final[f'parados_{cat}']
          + df_final[f'inactivos_{cat}'])
    diff = (suma - df_final[f'poblacion_mayor_16_{cat}']).abs()
    pct = (diff / df_final[f'poblacion_mayor_16_{cat}'] * 100)
    print(f'  {cat}: error medio = {diff.mean():.2f} ({pct.mean():.3f}%) | máx = {diff.max():.2f}')
```

Los chequeos de consistencia entre las propias series del ine revelan que la retropolación no fue realizada de forma perfecta. La suma total de población mayor de 16 años tiene un ligero error que será corregido recalculando la columna en base a las categorías que la componen de forma exhaustiva.

```python
# 1. Reemplazar parados con valores corregidos
columnas_parados = [c for c in df_paro.columns if c.startswith('parados_') and 'diferencia' not in c]

df_final = df_final.drop(columns=columnas_parados, errors='ignore')
df_final = df_final.merge(df_paro[['periodo'] + columnas_parados], on='periodo', how='left')

print(f'{len(columnas_parados)} columnas de parados reemplazadas')

# 2. Recalcular activos e inactivos por identidad contable
categorias = [
    'ambos_sexos_total', 'ambos_sexos_25_o_mas', 'ambos_sexos_menos_de_25',
    'hombres_total', 'hombres_25_o_mas', 'hombres_menos_de_25',
    'mujeres_total', 'mujeres_25_o_mas', 'mujeres_menos_de_25',
]

for cat in categorias:
    df_final[f'activos_{cat}'] = df_final[f'ocupados_{cat}'] + df_final[f'parados_{cat}']
    df_final[f'inactivos_{cat}'] = df_final[f'poblacion_mayor_16_{cat}'] - df_final[f'activos_{cat}']

print(f'{len(categorias)} categorías de activos e inactivos recalculadas por identidad contable')
```

# 5. Cálculo de nuevas variables

```python
# Tasas estándar del mercado laboral (definiciones INE/ILO/Eurostat)
# - Tasa de actividad   = activos    / población 16+   * 100
# - Tasa de empleo      = ocupados   / población 16+   * 100
# - Tasa de paro        = parados    / activos         * 100   (denominador OIT: fuerza laboral)
# - Tasa de inactividad = inactivos  / población 16+   * 100

categorias = [
    'ambos_sexos_total', 'ambos_sexos_25_o_mas', 'ambos_sexos_menos_de_25',
    'hombres_total',     'hombres_25_o_mas',     'hombres_menos_de_25',
    'mujeres_total',     'mujeres_25_o_mas',     'mujeres_menos_de_25',
]

for cat in categorias:
    df_final[f'tasa_actividad_{cat}']   = df_final[f'activos_{cat}']   / df_final[f'poblacion_mayor_16_{cat}'] * 100
    df_final[f'tasa_empleo_{cat}']      = df_final[f'ocupados_{cat}']  / df_final[f'poblacion_mayor_16_{cat}'] * 100
    df_final[f'tasa_paro_{cat}']        = df_final[f'parados_{cat}']   / df_final[f'activos_{cat}']            * 100
    df_final[f'tasa_inactividad_{cat}'] = df_final[f'inactivos_{cat}'] / df_final[f'poblacion_mayor_16_{cat}'] * 100

# Métricas derivadas

df_final['brecha_genero_actividad'] = df_final['tasa_actividad_hombres_total'] - df_final['tasa_actividad_mujeres_total']
df_final['brecha_genero_empleo']    = df_final['tasa_empleo_hombres_total']    - df_final['tasa_empleo_mujeres_total']
df_final['brecha_genero_paro']      = df_final['tasa_paro_mujeres_total']      - df_final['tasa_paro_hombres_total']

df_final.filter(like='tasa_').describe().round(2)
```

```python
# ── Añadir tasas de paro al CSV de parados ──────────────────────────────────
# Las tasas se calculan aquí porque este notebook tiene tanto parados como
# activos. Se guardan en el CSV de parados para análisis posterior.

from pathlib import Path

ruta_parados = Path(r'C:\Users\marco\PycharmProjects\TFM_Marcos\Datasets') / 'parados_sexo_edad.csv'
df_parados_csv = pd.read_csv(ruta_parados)
df_parados_csv['fecha'] = pd.to_datetime(df_parados_csv['fecha'])

# Columnas de tasa a añadir (9 tasas base + 2 derivadas)
cols_tasa = [
    'tasa_paro_ambos_sexos_total',
    'tasa_paro_ambos_sexos_25_o_mas',
    'tasa_paro_ambos_sexos_menos_de_25',
    'tasa_paro_hombres_total',
    'tasa_paro_hombres_25_o_mas',
    'tasa_paro_hombres_menos_de_25',
    'tasa_paro_mujeres_total',
    'tasa_paro_mujeres_25_o_mas',
    'tasa_paro_mujeres_menos_de_25',
    'brecha_genero_paro'
]

# Eliminar columnas de tasa si ya existen (por si se re-ejecuta)
cols_existentes = [c for c in cols_tasa if c in df_parados_csv.columns]
if cols_existentes:
    df_parados_csv = df_parados_csv.drop(columns=cols_existentes)
    print(f'Columnas de tasa previas eliminadas: {len(cols_existentes)}')

# Extraer tasas del df_final (solo fecha + columnas de tasa)
df_tasas = df_final[['fecha'] + cols_tasa].copy()

# Merge por fecha (left join: se conservan todas las fechas de parados)
df_parados_csv = df_parados_csv.merge(df_tasas, on='fecha', how='left')

# Guardar
df_parados_csv.to_csv(ruta_parados, index=False, encoding='utf-8-sig')
```
