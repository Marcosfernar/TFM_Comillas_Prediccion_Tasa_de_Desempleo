# 2. Limpieza

### Transformación del tipo de dato y limpieza inicial

```python
# Conversión de las variables categóricas a "category" por eficiencia
variables_categoricas = ['Sexo', 'Situación profesional']

for i, df in enumerate(dataframes):
    for variable in variables_categoricas:
        if variable in df.columns:
            df[variable] = df[variable].astype('category')
        else:
            continue
    dataframes[i] = df
```

```python
# Conversión de los datos numéricos a formato internacional, eliminar puntos y reemplazar las comas "," por puntos "."
for i, df in enumerate(dataframes):
    df['Total'] = df['Total'].astype('string')
    numeros_internacionales = df['Total'].str.replace('.', '', regex=False).str.replace(',', '.', regex=False)
    df['Total'] = pd.to_numeric(numeros_internacionales, errors='coerce')
    dataframes[i] = df
```

### Transformación de los datos de "Periodo"

```python
trimestres_map = {
    'I': '1',
    'II': '2',
    'III': '3',
    'IV': '4',
    '1': '1', '2': '2', '3': '3', '4': '4'
}

# Conversión de "Periodo" a string y a formato internacional para manejarlo mejor
for i, df in enumerate(dataframes):
    df['Periodo'] = df['Periodo'].astype('string')

    # Separación de la información en diferentes columnas para su posterior uso
    df['año'] = df['Periodo'].str[:4]
    df['trimestre'] = df['Periodo'].str[5:].map(trimestres_map)

    texto_fecha = df['año'].astype('string') + "Q" + df['trimestre'].astype('string')
    df['Periodo'] = texto_fecha.astype('category')
    df['fecha'] = pd.PeriodIndex(texto_fecha, freq="Q").to_timestamp()

    df['año'] = df['año'].astype(int)
    df['trimestre'] = df['trimestre'].astype(int)

    df = df.set_index('fecha').sort_index()
    df = df.reset_index()
    dataframes[i] = df

    print(f'Información acerca de {df.attrs['nombre']}')
    print(df.info())
```

Se aprecia como en la metodología más actualizada se excluyen las categorías  de "Ayuda en la empresa o negocio familiar", "Miembro de una cooperativa" y "Otra situación" por lo que se las excluirá para el cálculo de la nueva variable.
Además, revisando para el total de autónomos, se puede ver como el cambio es que estas categorías antes se consideraban asalariados y ahora se les considera empresarios. Así que vamos a calcular los totales y dropear las variables de con o sin empresarios pues son imoposibles de conseguir para las seires anteriores a 2002.

```python
def crear_total_asalariados(df):
    columnas_contexto = [columna for columna in df.columns if columna not in ['Situación profesional', 'Total']]

    # Filtramos las columnas según la metodología reciente
    mask = df['Situación profesional'].isin(['asalariados sector privado', 'asalariados sector público'])

    # Agrupamos por el resto de columnas que no contienen la información de la categoría o el total de ocupados
    df_total_asalariados = df[mask].groupby(columnas_contexto, as_index=False, observed=True)['Total'].sum()

    # Nombramos a las nuevas filas
    df_total_asalariados['Situación profesional'] = 'asalariados_total'

    # Unimos la nueva categoría al dataset original del que provenían los datos
    return pd.concat([df, df_total_asalariados], ignore_index=True)

df_76_95 = crear_total_asalariados(df_76_95)
df_96_01 = crear_total_asalariados(df_96_01)
```

```python
def crear_total_empresarios(df):
    columnas_contexto = [columna for columna in df.columns if columna not in ['Situación profesional', 'Total']]

    # Filtramos las columnas según la metodología reciente
    mask = df['Situación profesional'].isin(['empresario con o sin asalariados', 'ayuda familiar', 'miembro de cooperativa'])

    # Agrupamos por el resto de columnas que no contienen la información de la categoría o el total de ocupados
    df_total_asalariados = df[mask].groupby(columnas_contexto, as_index=False, observed=True)['Total'].sum()

    # Nombramos a las nuevas filas
    df_total_asalariados['Situación profesional'] = 'empresarios_total'

    # Unimos la nueva categoría al dataset original del que provenían los datos
    return pd.concat([df, df_total_asalariados], ignore_index=True)


df_76_95 = crear_total_empresarios(df_76_95)
df_96_01 = crear_total_empresarios(df_96_01)
df_02_10 = df_02_10[~df_02_10['Situación profesional'].isin(['Empresario sin asalariados o trabajador independiente', 'Empleador'])]
df_11_26 = df_11_26[~df_11_26['Situación profesional'].isin(['Empresario sin asalariados o trabajador independiente', 'Empleador'])]
```

# 3. Unión en un único dataset

```python
df_final= pd.concat([df_76_95, df_96_01, df_02_10, df_11_26], ignore_index=True)
# Eliminar duplicados de ruptura metodológica 2002, conservar serie más moderna
cols_clave = [c for c in df_final.columns if c not in ['Total', 'total', 'valor', 'Value']]
n_antes = len(df_final)
df_final = df_final.drop_duplicates(subset=cols_clave, keep='last').reset_index(drop=True)
print(f'Filas: {n_antes} → {len(df_final)} ({n_antes - len(df_final)} duplicados eliminados)')

sexo_map = {
    'Ambos sexos': 'ambos_sexos',
    'ambos sexos':'ambos_sexos',
    'Varones': 'hombres',
    'varones': 'hombres',
    'Hombres': 'hombres',
    'hombres': 'hombres',
    'Mujeres': 'mujeres',
    'mujeres': 'mujeres'
}

situacion_profesional_map = {
    'Total': 'total',
    'total': 'total',
    'empresarios_total': 'empresarios_total',
    'Trabajador por cuenta propia': 'empresarios_total',
    'Miembro de una cooperativa': 'miembro_cooperativa',
    'miembro de cooperativa': 'miembro_cooperativa',
    'Ayuda en la empresa o negocio familiar': 'ayuda_negocio_familiar',
    'ayuda familiar': 'ayuda_negocio_familiar',
    'Asalariados : Total': 'asalariados_total',
    'asalariados_total': 'asalariados_total',
    'Asalariado sector público': 'asalariados_sector_publico',
    'asalariados sector público': 'asalariados_sector_publico',
    'asalariados sector privado': 'asalariados_sector_privado',
    'Asalariado sector privado': 'asalariados_sector_privado',
    'Otra situación': 'ocupado_otra_situacion',
    'otra situación': 'ocupado_otra_situacion'

}

df_final['situacion_profesional'] = df_final['situacion_profesional'].map(situacion_profesional_map)

df_final['sexo'] = df_final['sexo'].map(sexo_map)

columnas_texto = df_final.select_dtypes(include=['category']).columns

for columna in columnas_texto:
    df_final[columna] = df_final[columna].str.lower()

df_final.head()
```

## Pivotando La tabla para poder analizarla y tener solo un registro por día

```python
df_final = df_final.pivot_table(
    index=['fecha', 'año', 'trimestre', 'periodo'],
    columns=['sexo', 'situacion_profesional'],
    values='valor'
).reset_index()


# Convirtiendo el nombre de las columnas en snake_case
df_final.columns = ['_'.join(columna).strip('_') if isinstance(columna, tuple) else columna for columna in df_final.columns.values]
df_final.columns = [columna.lower().replace(' ', '_') for columna in df_final.columns]

df_final.columns = [f'ocupados_{c}' if c not in ('fecha', 'año', 'trimestre', 'periodo') else c for c in df_final.columns]

df_final = df_final.sort_values('fecha').reset_index().drop(columns= 'index')
df_final
```
