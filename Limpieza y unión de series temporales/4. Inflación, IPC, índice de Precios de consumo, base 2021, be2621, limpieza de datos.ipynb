# 2. Limpieza

### Transformación de los datos de fecha

```python
df_54_25 = df_54_25.copy()
df_54_25['fecha'] = pd.to_datetime(df_54_25['fecha'])
df_54_25['año'] = df_54_25['fecha'].dt.year
df_54_25['trimestre'] = df_54_25['fecha'].dt.quarter
texto_fecha = df_54_25['año'].astype('string') + "Q" + df_54_25['trimestre'].astype('string')
df_54_25['periodo'] = texto_fecha

df_54_25 = df_54_25.set_index('fecha').sort_index()
df_54_25 = df_54_25.reset_index()

print(f'Información acerca de {df_54_25.attrs['nombre']}')
print(df_54_25.info())
```

# 3. Cálculo de factores de conversión e IPC en diferentes años base

```python
# Años base para los conversores
fechas = [1986, 2001, 2025]

# Calcular la media anual del IPC para cada año base
ipc_medias_base = {}

for año in fechas:
    datos_año = df_54_25[df_54_25['año'] == año]['IPC_2021']

    if len(datos_año) == 0:
        print(f'Año {año}: sin datos en el dataset. No se puede calcular el conversor.')
        continue

    media = datos_año.mean()
    n_meses = len(datos_año)
    ipc_medias_base[año] = media
    completitud = 'año sin nulos' if n_meses == 12 else f'{n_meses}/12 meses disponibles'


    print(f'Año {año}: IPC medio (base 2021) = {media:.4f} ({completitud})')
```

```python
# Calcular columnas de conversores
for año, ipc_media in ipc_medias_base.items():
    col_name = f'conversor_{año}'
    df_54_25[col_name] = (ipc_media / df_54_25['IPC_2021'])
    col_ipc = f'IPC_{año}'
    df_54_25[col_ipc] = (1/df_54_25[col_name]) * 100

print()
print('Columnas :')
print(df_54_25.columns.tolist())
```

```python
# Verificación del conversor_2025: en los meses de 20215debe ser ~ 1.0
check_2025 = df_54_25[df_54_25['año'] == 2025][['fecha', 'IPC_2021', 'IPC_2025', 'conversor_2025']]
print('Verificación conversor_2025 en meses de 2025 (debe ser ≈ 1.0):')
print(check_2025.to_string(index=False))
print(f'\nMedia conversor_2025 en 2025: {check_2025["conversor_2025"].mean():.6f} (debe ser ≈ 1.0)')
```

```python
check_2025 = df_54_25[df_54_25['año'] == 2025][['fecha', 'IPC_2021', 'IPC_2025', 'conversor_2025']]
print(check_2025.to_string(index=False))
```

```python
# Verificación: tabla resumen de deflactores en fechas clave
fechas_clave = ['1975-01-01',
                '1986-01-01',
                '1986-12-01',
                '1999-01-01',
                '2001-01-01',
                '2001-12-01',
                '2008-01-01',
                '2021-01-01',
                '2021-12-01',
                '2025-01-01',
                '2025-12-01'
                ]


cols_deflactor = ['fecha', ] + [f'IPC_{año}' for año in ipc_medias_base.keys()]
df_check = df_54_25[df_54_25['fecha'].isin(pd.to_datetime(fechas_clave))][cols_deflactor]

print('Comparación del IPC en fechas clave en diferentes bases:')
print(df_check.to_string(index=False))
```

# 4. Guardando el dataset final en un archivo

```python
# ── Recortar al rango de tasa de paro y guardar en Datasets ────────────
from pathlib import Path
FECHA_INICIO = '1974-07-01'
FECHA_FIN    = '2025-10-01'
ruta_datasets = Path(r'C:\Users\marco\PycharmProjects\TFM_Marcos\Datasets')

df_final = df_54_25.copy()


df_final = df_final[(df_final['fecha'] >= FECHA_INICIO) & (df_final['fecha'] <= FECHA_FIN)].copy()
df_final.to_csv(ruta_datasets / 'indice_precios_consumo_IPC_diferentes_bases.csv',
                index=False,
                encoding='utf-8-sig')
```
