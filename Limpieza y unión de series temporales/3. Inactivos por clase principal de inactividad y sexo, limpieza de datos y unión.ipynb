# 2. Limpieza

### Transformación del tipo de dato y limpieza inicial

```python
# Conversión de las variables categóricas a "category" por eficiencia
variables_categoricas = ['Sexo', 'sexo', 'Situación de inactividad', 'Clase principal de la inactividad']

for i, df in enumerate(dataframes):
    for variable in variables_categoricas:
        if variable in df.columns:
            df[variable] = df[variable].astype('category')
        else:
            continue
    dataframes[i] = df
```

```python
# Conversión de los datos numéricos a formato internacional, eliminar puntos y reemplazar las comas "," por puntos "."
for i, df in enumerate(dataframes):
    df['Total'] = df['Total'].astype('string')
    numeros_internacionales = df['Total'].str.replace('.', '', regex=False).str.replace(',', '.', regex=False)
    df['Total'] = pd.to_numeric(numeros_internacionales, errors='coerce')
    dataframes[i] = df
```

### Transformación de los datos de "Periodo"

```python
trimestres_map = {
    'I': '1',
    'II': '2',
    'III': '3',
    'IV': '4',
    '1': '1', '2': '2', '3': '3', '4': '4'
}

# Conversión de "Periodo" a string y a formato internacional para manejarlo mejor
for i, df in enumerate(dataframes):
    df['Periodo'] = df['Periodo'].astype('string')

    # Separación de la información en diferentes columnas para su posterior uso
    df['año'] = df['Periodo'].str[:4]
    df['trimestre'] = df['Periodo'].str[5:].map(trimestres_map)

    texto_fecha = df['año'].astype('string') + "Q" + df['trimestre'].astype('string')
    df['Periodo'] = texto_fecha.astype('category')
    df['fecha'] = pd.PeriodIndex(texto_fecha, freq="Q").to_timestamp()

    df['año'] = df['año'].astype(int)
    df['trimestre'] = df['trimestre'].astype(int)

    df = df.set_index('fecha').sort_index()
    df = df.reset_index()
    dataframes[i] = df

    print(f'Información acerca de {df.attrs['nombre']}')
    print(df.info())
```

### Homogeneización de las posibles situaciones de inactividad

```python
df_asalariados1 = df_76_95.copy()
df_asalariados2 = df_96_01.copy()
df_asalariados3 = df_02_26.copy()



# Para examinar el primer periodo de 1990 en el primer dataset (1976 - 1995)
df_1990 = df_asalariados1[(df_asalariados1['Sexo'] == 'ambos sexos') & (df_asalariados1['Periodo'] == '1990Q1')]

# Para examinar el primer periodo de 2000 en el segundo dataset (1996 - 2001)
df_2000 = df_asalariados2[(df_asalariados2['Sexo'] == 'ambos sexos') & (df_asalariados2['Periodo'] == '2000Q1')]

# Para examinar el primer periodo de 2010 en el tercer dataset (2002 - 2010)
df_2025 = df_asalariados3[(df_asalariados3['Sexo'] == 'Ambos sexos') & (df_asalariados3['Periodo'] == '2025Q1')]



columnas_a_borrar = ['Sexo', 'año', 'trimestre', 'fecha']
df_1990 = df_1990.drop(columns = columnas_a_borrar)
df_2000 = df_2000.drop(columns = columnas_a_borrar)
df_2025 = df_2025.drop(columns = columnas_a_borrar)
```

```python
columnas_contexto = [columna for columna in df_02_26.columns if columna not in ['Clase principal de la inactividad', 'Total']]

def unificar_inactividad(df):
    # Filtramos las columnas para adaptarnos a la metodología antigua con menos granularidad
    mask1 = df['Clase principal de la inactividad'].isin(['Jubilado o prejubilado', 'Perceptor de pensión distinta de la de jubilación o prejubilación'])
    mask2 = df['Clase principal de la inactividad'].isin(['Otras', 'Realizó sin remuneración trabajos sociales, actividades benéficas, etc.'])

    # Agrupamos por el resto de columnas que no contienen la información de la categoría o el total de parados
    df_pensionistas = df[mask1].groupby(columnas_contexto, as_index=False, observed=True)['Total'].sum()
    df_otras = df[mask2].groupby(columnas_contexto, as_index=False, observed=True)['Total'].sum()

    # Nombramos a las nuevas filas
    df_pensionistas['Clase principal de la inactividad'] = 'inactivos_pensionistas'
    df_otras['Clase principal de la inactividad'] = 'inactivos_otra_causa'

    # Eliminamos las filas granulares de la metodología reciente
    mask = mask1 | mask2
    df = df[~mask]

    # Unimos la nueva categoría al dataset original del que provenían los datos
    return pd.concat([df, df_pensionistas, df_otras], ignore_index=True)


df_02_26 = unificar_inactividad(df_02_26)

```

# 3. Unión en un único dataset

```python
df_final= pd.concat([df_76_95, df_96_01, df_02_26], ignore_index=True)
# Eliminar duplicados de ruptura metodológica 2002, conservar serie más moderna
cols_clave = [c for c in df_final.columns if c not in ['Total', 'total', 'valor', 'Value']]
n_antes = len(df_final)
df_final = df_final.drop_duplicates(subset=cols_clave, keep='last').reset_index(drop=True)
print(f'Filas: {n_antes} → {len(df_final)} ({n_antes - len(df_final)} duplicados eliminados)')

sexo_map = {
    'Ambos sexos': 'ambos_sexos',
    'ambos sexos':'ambos_sexos',
    'Varones': 'hombres',
    'varones': 'hombres',
    'Hombres': 'hombres',
    'hombres': 'hombres',
    'Mujeres': 'mujeres',
    'mujeres': 'mujeres'
}

clase_principal_inactividad_map = {
    'Total': 'total',
    'total': 'total',
    'Estudiante': 'estudiantes',
    'estudiante': 'estudiantes',
    'inactivos_pensionistas': 'pensionistas',
    'jubilado o pensionista': 'pensionistas',
    'Jubilado o pensionista': 'pensionistas',
    'labores del hogar': 'labores_hogar',
    'Labores del hogar': 'labores_hogar',
    'incapacidad permanente': 'incapacidad_permanente',
    'Incapacidad permanente': 'incapacidad_permanente',
    'Incapacitado permanente': 'incapacidad_permanente',
    'Otra': 'otra_causa',
    'otra': 'otra_causa',
    'inactivos_otra_causa': 'otra_causa',
}

df_final['situacion_inactividad'] = df_final['situacion_inactividad'].map(clase_principal_inactividad_map)

df_final['sexo'] = df_final['sexo'].map(sexo_map)

columnas_texto = df_final.select_dtypes(include=['category']).columns

for columna in columnas_texto:
    df_final[columna] = df_final[columna].str.lower()

df_final.head()
```

## Pivotando La tabla para poder analizarla y tener solo un registro por día

```python
df_final = df_final.pivot_table(
    index=['fecha', 'año', 'trimestre', 'periodo'],
    columns=['sexo', 'situacion_inactividad'],
    values='valor'
).reset_index()


# Convirtiendo el nombre de las columnas en snake_case
df_final.columns = ['_'.join(columna).strip('_') if isinstance(columna, tuple) else columna for columna in df_final.columns.values]
df_final.columns = [columna.lower().replace(' ', '_') for columna in df_final.columns]

df_final.columns = [f'inactivos_{c}' if c not in ('fecha', 'año', 'trimestre', 'periodo') else c for c in df_final.columns]

df_final = df_final.sort_values('fecha').reset_index().drop(columns= 'index')
df_final
```

# 4. Ajustando los datos anteriores a 2002 gracias a la serie retropolada de parados
La diferencia entre la metodología antigua y la moderna pasará a inactivos (otra causa)

```python
from pathlib import Path
ruta_datasets = Path(r'C:\Users\marco\PycharmProjects\TFM_Marcos\Datasets')

df_retropol = pd.read_csv(ruta_datasets / 'poblacion_relacion_actividad_economica_sexo_edad.csv')
df_retropol = df_retropol.filter(regex='^(periodo|inactivos_.*_total)$')
df_raw = df_final.filter(regex='^(periodo|inactivos_(ambos_sexos|hombres|mujeres)_(total|otra_causa))')
df_intermedio = df_raw.merge(df_retropol, on='periodo', how='left', suffixes=('_raw', '_retropol'))
df_intermedio
```

```python
diferencia_ambos_sexos = df_intermedio['inactivos_ambos_sexos_total_retropol'] - df_intermedio['inactivos_ambos_sexos_total_raw']
diferencia_hombres = df_intermedio['inactivos_hombres_total_retropol'] - df_intermedio['inactivos_hombres_total_raw']
diferencia_mujeres = df_intermedio['inactivos_mujeres_total_retropol'] - df_intermedio['inactivos_mujeres_total_raw']

# Total de Inactivos
df_final['inactivos_ambos_sexos_total'] += diferencia_ambos_sexos
df_final['inactivos_hombres_total'] += diferencia_hombres
df_final['inactivos_mujeres_total'] += diferencia_mujeres

# Inactivos, otra causa
df_final['inactivos_ambos_sexos_otra_causa'] += diferencia_ambos_sexos
df_final['inactivos_hombres_otra_causa'] += diferencia_hombres
df_final['inactivos_mujeres_otra_causa'] += diferencia_mujeres
```
